```python
from datasets import load_dataset
import pandas as pd
```

```python
ds = load_dataset("AI4Sec/cti-bench", "cti-mcq")
data = ds['test']
# Basic information
print(f"Number of examples: {len(data)}")
print(f"Column names: {data.column_names}")
print(f"Features: {data.features}")
print()

# Show first example
print("First example:")
print(data[0])

num_data = len(data)
data_subset = data.select(range(num_data))
```

```python
prompt_data = data_subset["Prompt"]
prompt_data[0]
```

```python
prompt_data = list(data_subset["Prompt"])

for i in range(len(prompt_data)):
    prompt_data[i] = prompt_data[i].split('**Question:** ', 1)[1]

prompt_data[0]
```

```python
data = {"Prompt": prompt_data, "GT": data_subset["GT"]}
df = pd.DataFrame(data)


df.to_csv("evaluation_dataset_splitted.csv")
```
