```python
# Load dependencies
# from transformers import AutoTokenizer, AutoModelForCausalLM
# import torch
# from datasets import load_from_disk
from datasets import load_dataset
```

```python
#load evaluation from huggingface
ds = load_dataset("AI4Sec/cti-bench", "cti-mcq")
data = ds['test'] if 'test' in ds else ds['train']
# Basic information
print(f"Number of examples: {len(data)}")
print(f"Column names: {data.column_names}")
print(f"Features: {data.features}")
print()

# Show first example
print("First example:")
print(data[0])

num_data = len(data)
data_subset = data.select(range(num_data))
```

```python
# takes the ground truth
gt = data_subset["GT"]
```

```python
# takes the data from the files
filename_base = "./answer_mcq/cti-mcq_qwen-2-base_first2500_result.txt"
with open(filename_base, 'r') as f:
    content = f.read()
    ans = content.split('\n')

print(ans)
```

```python
# takes the data from the files
filename_finetuned = "./answer_mcq/cti-mcq_qwen-2_first2500_result.txt"
with open(filename_finetuned, 'r') as f:
    content = f.read()
    ans_ft = content.split('\n')

print(ans_ft)
```

```python
filename_datafinetuned = "./answer_mcq/cti-mcq_qwen-2-datafinetuned_first2500_result.txt"
with open(filename_datafinetuned, 'r') as f:
    content = f.read()
    ans_dft = content.split('\n')

print(ans_dft)
```

# Result Before Finetuned

```python
correct = 0
for i in range(len(gt)):
    if gt[i] == ans[i]:
        correct += 1

print(correct)

print(correct/2500*100)
```

# Result after finetuned

```python
correct = 0
for i in range(len(gt)):
    if gt[i] == ans_ft[i]:
        correct += 1

print(correct)

print(correct/2500*100)
```

# Result after datafinetuned (adding system prompt)

```python
correct = 0
for i in range(len(gt)):
    if gt[i] == ans_dft[i]:
        correct += 1

print(correct)

print(correct/2500*100)
```
